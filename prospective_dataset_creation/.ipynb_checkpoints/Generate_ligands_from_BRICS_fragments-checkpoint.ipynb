```python
def recursive_trim(orig_smi):
    with suppress_stdout_stderr():
        chunks,unmatched = rdRGD.RGroupDecompose(
                  BRICS_terminals_as_mols,
                  [ Chem.MolFromSmiles(orig_smi) ],
                  asSmiles=True, options=RGDparams )
    if(len(unmatched)==1):
        return(orig_smi)
    else:
#         print("left:", chunks)
        for d in chunks:
            smi=""
            for key in d:
                if key!="Core" and "[Au]" in d[key]:
                    smi=d[key]
                    break;
            #replace term with Ag
            smi = Chem.CanonSmiles(re.sub(r'\[\*:[0-9]+\]', '[Ag]', smi) )
            return(recursive_trim(smi))

linkers_trimmed=[]
for smi in linkers_for_triming:
    trimmed=recursive_trim(smi)
    if trimmed not in linkers_trimmed:
        linkers_trimmed.append(trimmed)
#     print(smi, "\t\t", trimmed)

# linkers_trimmed=[recursive_trim(smi) for smi in linkers_for_triming]
```

```python
def convert_fragment(smi):
    mol = Chem.MolFromSmiles(smi)
    smi = re.sub(r'\@', '', smi) # remove steric character of bonds
    smi = re.sub(r'\[CH\]', 'C', smi) # remove explicit specification of aliphatic carbon's hydrogens
    patt_smi = re.sub(r'\*n([0-9]?)', '[nH]'+r'\1', smi) # for termini
    patt_smi=re.sub(r'\*|\(\*\)', '', patt_smi) # for termini
    patt_smi=re.sub(r'\[Ag\]|\(\[Ag\]\)', '', patt_smi) # for fragment to terminus connections
    patt_smi=re.sub(r'\[Au\]|\(\[Au\]\)', '', patt_smi) # for fragment to scaffold connections
    
#     print(smi, patt_smi)
    
    patt = Chem.MolFromSmiles(patt_smi)
    hit_ats = list(mol.GetSubstructMatch(patt))
    if(len(hit_ats)==0):
        raise(Exception("No substructure match found!"+f"\toriginal: {smi}\tw/o R-groups: {patt_smi}"))

    for atom in patt.GetAtoms():
        atom.SetAtomMapNum(atom.GetIdx())

    for v,i in enumerate(hit_ats):
        mol.GetAtoms()[i].SetAtomMapNum(v)

    # which atoms are bonded to Ag & Au?
    open_slots=[]
    for b in mol.GetBonds():
        a1=mol.GetAtoms()[b.GetBeginAtomIdx()]
        a2=mol.GetAtoms()[b.GetEndAtomIdx()]

        a1e=a1.GetAtomicNum()
        a2e=a2.GetAtomicNum()

        # Ag -> 47; Au -> 79; * -> 0
        if(a1e == 47 or a1e == 79 or a1e == 0):
            open_slots.append(a2.GetAtomMapNum())
        elif(a2e == 47 or a2e == 79 or a2e == 0):
            open_slots.append(a1.GetAtomMapNum())

#     print(patt_smi, open_slots)
    Draw.MolsToGridImage([mol, patt],
                     subImgSize=(500, 500), molsPerRow=3)
    return(patt_smi, open_slots)


# convert termini
ready_terminals=[]
for term in BRICS_terminals:
    ready_terminals.append(convert_fragment(term))
    
# convert linkers
ready_linkers=[]
for l in linkers_trimmed:
    ready_linkers.append(convert_fragment(l))

# ready_linkers=[]
# linkers_canonical_smi=[]
# linkers_smi=[]
# for l in linkers_trimmed:
#     smi, open_slots = convert_fragment(l)
#     canonical_smi = Chem.CanonSmiles(smi)
#     if(canonical_smi in linkers_canonical_smi):
#         if(smi not in linkers_smi):
#             raise(Exception("reordered linker duplicate detected"));
            
#     if(smi in linkers_smi):
#         i=linkers_smi.index(smi)
#         for s in open_slots:
#             if open_slots.count(s) == 1: # happens only once
#                 if(s not in ready_linkers[i][1]):
#                     ready_linkers[i][1].append(s) # add the open slot
#             else:
#                 raise(Exception("multiple open slot on the same atom and they aren't present in ready_linkers yet"));
#     else:
#         ready_linkers.append([smi, open_slots])
#         linkers_smi.append(smi)
#         linkers_canonical_smi.append(canonical_smi)
```

```python
def make_bonded(scaf, frag, scaf_atom_id, frag_atom_id, scaf_free, frag_free):
    len_scaffold=len(scaf.GetAtoms())
    try:
        combo = Chem.CombineMols(scaf, frag)
    except ArgumentError as e:
        print("problem with fragment {}".format(frag))
        raise(e)
    ats_ndx = None
    for i in frag_free:
        if(i==frag_atom_id):
            ats_ndx=i
            break;
    if(ats_ndx is None):
        raise(Exception("Fragment atom to bond is not free to do so."))
        
    edcombo = Chem.EditableMol(combo)
    edcombo.AddBond(scaf_atom_id, frag_atom_id+len_scaffold, order=Chem.rdchem.BondType.SINGLE)
    back = edcombo.GetMol()
    #remove implicit hydrogens that can't be corrected for when a bond is added
    back=Chem.rdmolops.RemoveHs(back, implicitOnly=True)

    with suppress_stdout_stderr():
        problems = Chem.DetectChemistryProblems(back)
    for p in problems:
        if(p.GetType()=="AtomValenceException"):
            nH=back.GetAtoms()[p.GetAtomIdx()].GetNumExplicitHs()
            back.GetAtoms()[p.GetAtomIdx()].SetNumExplicitHs(nH-1)
    try:
        Chem.SanitizeMol(back)
    except Exception as e:
        print("fragment:", Chem.MolToSmiles(frag), f"\nattaching by fragment atom {frag_atom_id} or {frag_atom_id+len_scaffold} of comb to atom {scaf_atom_id} of curerent scaffold.")
        raise(e)

    new_free = [s for s in scaf_free if s!=scaf_atom_id] + [s+len_scaffold for j,s in enumerate(frag_free) if j!=ats_ndx]
    return(back, new_free)

# def expand_w_fragments(scaffold, fragments, currently_free):
#     tests=[]
#     tests_unique_SMILES=[]
#     tests_free=[]

#     for f in list(set(currently_free)): #only do this for unique free slots, avoids double processing of linkers' sites
#         for i,frag in enumerate(fragments):
#             m_frag=Chem.MolFromSmiles(frag[0])
#             for ats in frag[1]:
#                 back, new_free = make_bonded(scaffold, m_frag, f, ats, currently_free, frag[1])
                    
#                 smiles=Chem.MolToSmiles(back, kekuleSmiles=False)
#                 smiles=Chem.CanonSmiles(smiles)

#                 if(not smiles in tests_unique_SMILES): # ensure uniqueness
#                     tests_unique_SMILES.append(smiles)
#                     tests.append(back)
#                     tests_free.append(new_free)
                    
#     return(tests, tests_free)
    

def expand_w_fragments(scaffold, fragments, currently_free):
    tests=[]
#     tests_unique_SMILES=[]
    tests_free=[]
    
    unique_reprs=[]

    for f in list(set(currently_free)): #only do this for unique free slots, avoids double processing of linkers' sites
        for i,frag in enumerate(fragments):
            m_frag=Chem.MolFromSmiles(frag[0])
            len_scaffold=len(scaffold.GetAtoms())
            try:
                combo = Chem.CombineMols(scaffold, m_frag)
            except ArgumentError as e:
                print("problem with fragment {}".format(frag))
                raise(e)
            for ats_ndx, ats in enumerate(frag[1]):
                edcombo = Chem.EditableMol(combo)
                edcombo.AddBond(f, ats+len_scaffold, order=Chem.rdchem.BondType.SINGLE)
                back = edcombo.GetMol()
                #remove implicit hydrogens that can't be corrected for when a bond is added
                back=Chem.rdmolops.RemoveHs(back, implicitOnly=True)
                
                with suppress_stdout_stderr():
                    problems = Chem.DetectChemistryProblems(back)
                for p in problems:
                    if(p.GetType()=="AtomValenceException"):
                        nH=back.GetAtoms()[p.GetAtomIdx()].GetNumExplicitHs()
                        back.GetAtoms()[p.GetAtomIdx()].SetNumExplicitHs(nH-1)
#                 Chem.SanitizeMol(back)
                
                try:
                    Chem.SanitizeMol(back)
                except Exception as e:
                    print("fragment:", frag[0], f"\nattaching by fragment atom {ats} or {ats+len_scaffold} of comb to atom {f} of curerent scaffold.")
                    raise(e)
#                     if("Explicit valence" in str(e)):
#                         try:
#                             nH=back.GetAtoms()[ats+len_scaffold].GetNumExplicitHs()
#                             back.GetAtoms()[ats+len_scaffold].SetNumExplicitHs(nH-1)
#                             Chem.SanitizeMol(back)
#                         except Exception as e:
#                             print("Failed handling too high valance exception")
#                             raise(e)
#                     else:    
#                         #print(ats+len_scaffold,frag)
#                         #print(ats,frag)
#                         #print(f, currently_free)
#                         #Draw.MolsToGridImage([m_frag])
#                         #return(tests, tests_free)
#                         raise(e)
                    
                    
                smiles=Chem.MolToSmiles(back, kekuleSmiles=False)
                smiles=Chem.CanonSmiles(smiles)

#                 if(not smiles in tests_unique_SMILES): # ensure uniqueness
#                     tests_unique_SMILES.append(smiles)
#                     tests.append(back)
#                     #find new free sites
#                     new_free = [s for s in currently_free if s!=f] + [s+len_scaffold for j,s in enumerate(frag[1]) if j!=ats_ndx]
#                     tests_free.append(new_free)
                    
                new_free = [s for s in currently_free if s!=f] + [s+len_scaffold for j,s in enumerate(frag[1]) if j!=ats_ndx]
                tup=(smiles, new_free)
                if(tup not in unique_reprs):
                    unique_reprs.append(tup)
                    tests.append(back)
                    tests_free.append(new_free)
                    
                    
    return(tests, tests_free)
```

```python
# the code is borrowed from following url.
# https://bit.ly/3aTUBCU
class HorizontalDisplay:
    def __init__(self, *args):
        self.args = args
 
    def _repr_html_(self):
        template = '<div style="float:left;padding:10px;">{0}</div>'
        return "\n".join(template.format(arg)
                         for arg in self.args)
```

```python
def Custom_MolsToGridImage(mols, molsPerRow=3, subImgSize=(200, 200), legends=None,
                     highlightAtomLists=None, highlightBondLists=None, drawOptions=None,
                     returnPNG=False, **kwargs):
    """ returns a PIL Image of the grid
    """
    if legends is None:
        legends = [''] * len(mols)

    nRows = len(mols) // molsPerRow
    if len(mols) % molsPerRow:
        nRows += 1

    fullSize = (molsPerRow * subImgSize[0], nRows * subImgSize[1])
    #d2d = rdMolDraw2D.MolDraw2DCairo(fullSize[0], fullSize[1], subImgSize[0], subImgSize[1])
    d2d = rdMolDraw2D.MolDraw2DSVG(fullSize[0], fullSize[1], subImgSize[0], subImgSize[1])
    if drawOptions is not None:
        d2d.SetDrawOptions(drawOptions)
    else:
        dops = d2d.drawOptions()
        for k, v in list(kwargs.items()):
            #if hasattr(dops, k):
            setattr(dops, k, v)
            del kwargs[k]

    d2d.DrawMolecules(list(mols), legends=legends or None, highlightAtoms=highlightAtomLists,
                      highlightBonds=highlightBondLists, **kwargs)
    d2d.FinishDrawing()
    res = d2d.GetDrawingText()
    return(res)
```
